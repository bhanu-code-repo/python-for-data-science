```python
# importing packages
import pandas as pd
```

```python
# create dictionary - a python structure which stores key value pair
property_data = { 'id': ['10001','10002','10003','10004','10005','10006','10007','10008','10009'],
                  'name': ['SAIDEEP', 'SHOBHA', 'BREN', 'MANTRI', 'BRIGADE', 'PRESTIGE', 'ASSETZ', 'SPRING FIELDS', 'DEFINER'],
                  'occupied': ['Y','N','Y','Y','N','Y','Y','N','N'],
                  'bhk_type': [3,3,2,1,2,3,3,2,2],
                  'area_sqft':[1700,1400,1250,870,1140,2100,1500,980,1120]
                }
```

here keys are `id, name, occupied, bhk_type, are_sqft` and values are their corresponding data i.e. `id values are 10001, 10002, ...`

```python
# create dataframe
df = pd.DataFrame(property_data)
```

```python
# view dataframe
df
```

dataframe has three components - index `0,1,2,...,8`, column `id, name, occupied, bhk_type, are_sqft` and data (all column values)

```python
# reading single column - option 1
df['name']
```

```python
# reading single column - option 2
df.name
```

```python
# reading single column - option 3
prop_names = df['name']
prop_names
```

```python
# creating new column in dataframe
df['status'] = df['name'] + '_' + df['occupied']
```

```python
# view dataframe
df
```

as we can see new column `status` is added to the existing dataframe

```python
# data selection/sorting - get properties of type greater than 2 BHK
interested_properties = df[df['bhk_type'] > 2]
interested_properties
```

```python
# data selection/sorting - get properties of type 2 BHK
interested_properties = df[df['bhk_type'] == 2]
interested_properties
```

```python
# data selection/sorting - get properties of type less tha 4 BHK
interested_properties = df[df['bhk_type'] < 4]
interested_properties
```

```python
# data selection/sorting - get properties of type 3 BHK and are occupied
interested_properties = df[(df['bhk_type'] == 3) & (df['occupied'] == 'Y')]
interested_properties
```

```python
# data selection/sorting - get properties of type 3 BHK or are occupied
interested_properties = df[(df['bhk_type'] == 3) | (df['occupied'] == 'Y')]
interested_properties
```
