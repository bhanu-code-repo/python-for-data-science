```python
# importing packages
import pandas as pd
```

## Import CSV with `pd.read_csv()` function

```python
# reading dataset - csv file (option 1)
df = pd.read_csv('datasets/world-happiness-report-2021.csv')
df
```

```python
# reading dataset - csv file (option 2)
# Note - if we try to read this csv without encoding, we will get this error
# 'utf-8' codec can't decode byte 0xe9 in position 2776: invalid continuation byte
# for fixing such error we need to set encoding type
df = pd.read_csv('datasets/asian-restaurants.csv', encoding='latin-1')
df
```

## Explore Data

```python
# head function - by default select top 5 rows
df.head()
```

### explore `tail()` function

```python
# tail function - by default select last 5 rows
df.tail()
```

### Sunsetting Data - `loc` and `iloc`

```python
# select rows by index (note - index begins from 0)
df.loc[7:14]
```

```python
# select single cell value
df.loc[10, 'restaurant']
```

```python
# select single cell by indexing row and column value
# row index - 10, column index - 1
df.iloc[10,1]
```
